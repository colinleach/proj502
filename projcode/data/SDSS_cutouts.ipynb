```python
coord = sdss.coords[1]
ra = coord[1]
dec = coord[2]
sdss_url = "http://skyserver.sdss.org/dr14/SkyServerWS/ImgCutout/getjpeg"
```

```python
def get_img(jpeg_size):
    img_url = sdss_url + f"?ra={ra:.5f}&dec={dec:.5f}&width={jpeg_size}&height={jpeg_size}"
    return requests.get(img_url, stream=True)
```

```python
img_data = get_img(424)
img = Image.open(img_data.raw)

plt.imshow(img);
```
